```python
import requests as rq
from bs4 import BeautifulSoup
import pprint


final_marvel_movie_list = []
temp = ""
res = rq.get("https://en.wikipedia.org/wiki/List_of_Marvel_Cinematic_Universe_films")                                                  
soup = BeautifulSoup(res.text, 'html.parser') 
table = soup.find("table",class_="wikitable")
for items in table.find_all("tr"):
    data = [' '.join(item.text.split()) for item in items.find_all(['th','td'])]
    if len(data) > 2:
        temp_data = data[0] + " | " +  data[1]
    final_marvel_movie_list.append(temp_data)

for index,items in enumerate(final_marvel_movie_list):
    if index > 1:
        pprint.pprint(items)
        
        

final_text = "\n\n".join([names for names in final_marvel_movie_list])
with open("movies1.txt", "w+") as file:
    file.write(final_text)
        
```
